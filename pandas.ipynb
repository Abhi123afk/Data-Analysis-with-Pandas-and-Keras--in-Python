# Practice Problem (multivariate linear regression)

```python
from sklearn import linear_model
import pandas as pd
```

```python
df=pd.read_csv('demoreg.csv')
df
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['area','rooms','age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,30]])
```
